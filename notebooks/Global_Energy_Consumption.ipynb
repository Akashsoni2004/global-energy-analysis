# 6. Feature Engineering

### 6.1 Renewable Percentage Column

```python
df["renewable_percentage"] = ((df["renewable_share"] / df["energy_consumption"]) * 100)
```

### 6.2 Energy Per Capita

```python
df["energy_per_capita"] = (df["energy_consumption"] / df["urban_population"])
```

### 6.3 CO<sub>2</sub> Per Capita

```python
df["co2_per_capita"] = (df["co2_emission"] / df["urban_population"])
```
