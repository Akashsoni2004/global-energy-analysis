# Introduction
__Global Energy Consumption__ has increased rapidly over the past few years due to industrilization, population growth, urban developement, and technological advancement. The Excessive dependence on fossil fuels such as coal, oil, and natural gas has led to serious environmental challenges including climate change, global warming, and increasig carbon dioxide (CO<sub>2</sub>) emissions.

The dataset provides __five years (2020-2024)__ of data combining __global energy consumption__ statistics for __50__ countries across all continents.

### The project aims to:
- identify __consumption patterns__
- compare __fossil fuels__ and __renewable energy__ usage
- study __(CO<sub>2</sub>) emissions trends__
- evaluate __energy efficiency__ across different countries and years.

By analyzing these factors, the project provides actionable insights and recommendations that can help in promoting renewable energy adoption, improving energy efficiency, and supporting sustainable developement goals.
