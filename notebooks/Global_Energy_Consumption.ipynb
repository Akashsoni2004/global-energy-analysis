# 6. Feature Engineering

### 6.1 Non-Renewable consuption share Column

```python
df["non_renewable_share"] = (100 - df["renewable_share"])
```

### 6.2 Energy Per Capita

```python
df["energy_per_capita"] = (df["energy_consumption"] / df["urban_population"])
```

### 6.3 CO<sub>2</sub> Per Capita

```python
df["co2_per_capita"] = (df["co2_emission"] / df["urban_population"])
```

### 6.4 Year column

```python
df["year"]=pd.to_datetime(df['date']).dt.year
```

# 7. Exploratory Data Analysis (EDA)

### 7.1 Top Energy Consuming Countries

```python
country_energy= df.groupby('country')['energy_consumption'].sum().sort_values(ascending=False)
sns.barplot(x=country_energy.values, y=country_energy.index, palette="coolwarm")


plt.title("Energy consumption by country")
plt.xlabel("Energy Consumption")
plt.ylabel("Country")
plt.show()
```

### 7.2 Renewable Energy Trend Over Time

```python
sns.lineplot(data=df, x="year", y="renewable_share")
plt.title('Renewable Energy Trend Over Time')
plt.xlabel('Year')
plt.ylabel('Renewable Energy')
plt.show()
```

### 7.3 CO<sub>2</sub> Emission Trend

```python
sns.lineplot(data=df, x="year", y="co2_emission", markers="o")
plt.title('Average CO₂ Emission Trend')
plt.xlabel('Year')
plt.ylabel('CO₂ Emission')
plt.show()
```

### 7.4 Correlation Heatmap

```python
plt.figure(figsize=(14,10))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()
```
